```python
from datetime import datetime

import numpy as np
import matplotlib.pyplot as plt
import h5py

# make graphics inline
%matplotlib inline

from SharpReader import SharpReader
from SharpWriter import SharpWriter
```

# Define experiment-specific parameters

```python
sid = 27223
fields = ['zpssx','zpssy']

sharpReader = SharpReader()
# x_c, y_c, xn, yn, threshold
sharpReader.init(64, 147, 128, 128, 2)

sharpWriter = SharpWriter()
# pixel size (um), distance (m), wavelength (nm), det_side 
sharpWriter.init(55, 0.64, 0.1331, 128) 

prbfile = '../../data/27223/recon_21678_t4_probe_ave_rp.npy'

cxifile = '../../data/27223/hxn27223.cxi'
```

# Get metadata from databroker

```python
print("getting fnames, points from db ...");
t1 = datetime.now();
fnames, ic = sharpReader.get_merlin1_fnames(sid)
xs, ys = sharpReader.get_points(sid, fields)
t2 = datetime.now();
print ("processing time: ", (t2 - t1), ", fnames: ", len(fnames), ", ic: ", len(ic), ", x: ", len(xs), ", y: ", len(ys));
```

# Initialize Spark

```python
from pyspark.sql import SparkSession

print ("spark initialization ...");
t1 = datetime.now();
spark = SparkSession.builder.appName("db2sharp.spark.17554").getOrCreate()
t2 = datetime.now();
print ("processing  time: ", (t2 - t1));
```

# Load files

```python
def load_files(kvs):
    
    from SharpReader import SharpReader
    sharpReader = SharpReader()
    # x_c, y_c, xn, yn, threshold
    sharpReader.init(60, 66, 100, 100, 2)   
    frames = sharpReader.load_files(kvs['sid'], kvs['fnames'], kvs['ic'])
    
    frmfile = './frames.' + str(kvs['sid']) + '.' + str(kvs['pid']) + '.npy'
    np.save(frmfile, frames)
    return frmfile


partitions = 20

print("loading files ...");
t1 = datetime.now();

pfnames = np.split(fnames, partitions)
pics    = np.split(ic, partitions)

env = []
for i in range(0, partitions):
    kvs = {'pid' : i, 'sid' : sid, 'fnames' : pfnames[i], 'ic' : pics[i]}
    env.append(kvs)
    
rdd = spark.sparkContext.parallelize(env, partitions)
frmfiles = rdd.map(load_files).collect()

t2 = datetime.now();
print ("processing time: ", (t2 - t1))

for i in range(0, partitions):
    print(i, ", frmfile: ", frmfiles[i])
```

# Combine files saved by the Spark workers

```python
print("loading and combining the Spark files ...");
t1 = datetime.now();

aframes = []
for frmfile in frmfiles:
    pframes = np.load(frmfile)
    aframes.append(pframes)   
frames = np.concatenate(aframes)
    
t2 = datetime.now();
print ("processing time: ", (t2 - t1), "frames: ", len(frames))
```

# Write data to the SHARP-NSLS2 input file 

```python
print("write cxi file ...");
t1 = datetime.now();
emptyFrames = sharpWriter.set_frames(frames)
X, Y = sharpWriter.set_points(xs, ys)
prb = sharpWriter.set_probe(prbfile)
sharpWriter.write(cxifile)
t2 = datetime.now();
print ("processing time: ", (t2 - t1))
print ("emptyFrames: ", len(emptyFrames));
print ("X and Y, min/max/size ", min(X), max(X), X.size, min(Y), max(Y), Y.size);
```
